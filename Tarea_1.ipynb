```python
#from sklearn.linear_model import
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

Modelo_LR = LogisticRegression(random_state=1)
Modelo_DT = DecisionTreeClassifier(random_state=1)
```

```python
Modelo_LR.fit(csv_train[csv_trainX],csv_train[csv_trainY])
Modelo_DT.fit(csv_train[csv_trainX],csv_train[csv_trainY])
```

```python
csv_test.columns[:132]
```

```python
pred_DT = Modelo_DT.predict(csv_test[csv_test.columns[:132]])
pred_LR = Modelo_LR.predict(csv_test[csv_test.columns[:132]])
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import balanced_accuracy_score
print(accuracy_score(pred_LR,csv_test[csv_trainY]))
print(accuracy_score(pred_DT,csv_test[csv_trainY]))
print(balanced_accuracy_score(pred_LR,csv_test[csv_trainY]))
print(balanced_accuracy_score(pred_DT,csv_test[csv_trainY]))
```
